```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(filename="images/sudoku.jpg", flags=0)
img = np.float32(img)
plt.imshow(X=img, cmap="gray");
```

```python
# Harris Corner Detection
harris = cv2.cornerHarris(src=img, blockSize=3, ksize=3, k=0.04)
plt.imshow(X=harris, cmap="gray");
```

```python
# Shi Tomsai Detection
corners = cv2.goodFeaturesToTrack(image=img, maxCorners=100, qualityLevel=0.01, minDistance=10)
corners = np.int64(corners)

for element in corners:
    x, y = element.ravel()
    cv2.circle(img=img,
               center=(x, y),
               radius=3,
               color=(125, 125, 125),
               thickness=cv2.FILLED)
plt.imshow(X=img, cmap="gray");
```

## cv2.cornerHarris(src, blockSize, ksize, k, dst, borderType)
### Harris ke alglama yntemini kullanarak bir grntdeki keleri tespit etmek iin kullanlan bir OpenCV fonksiyonudur. Bu yntem, kelerin kontrast, younluk ve dorultu gibi zelliklerine dayanarak keleri belirler.
### src: Giri grnts
### blockSize: Ke tespiti iin kullanlacak piksel blounun boyutu. Bu blok iindeki gradyanlarn hesaplanaca alan belirler. Genellikle 2x2 veya 3x3 boyutunda bir deer kullanlr.
### ksize: Sobel gradyan hesaplama iin kullanlacak Sobel ekirdeinin boyutu. Genellikle 3x3 boyutunda bir deer kullanlr.
### k: Harris ke alglama fonksiyonunda kullanlan serbest parametre. Bu deer, ke noktalarnn belirlenmesinde nemli bir rol oynar. Tipi olarak 0.04 ile 0.06 arasnda bir deer kullanlr.

## cv2.goodFeaturesToTrack(image, maxCorners, qualityLevel, minDistance, corners, mask, blocksize, useHarrisDetector, k)
### Bir grntdeki iyi takip edilebilir zelliklerin (genellikle kelerin) tespit edilmesi iin kullanlan bir OpenCV fonksiyonudur. Bu fonksiyon, Shi-Tomasi ke alglama yntemini kullanr.
### image: Giri grnts.
### maxCorners: Tespit edilecek en fazla ke says, Eer tespit edilen ke says bu deeri aarsa, en gl keler seilir.
### qualityLevel: Ke niteliini belirleyen eik deeri. Bu deer, kelerin ne kadar iyi olduunu kontrol eder. Tipik olarak 0.01 ile 0.1 arasnda bir deer kllanr.
### minDistance: Tespit edilen keler arasndaki minimum mesafe. Bu parametre, tespit edilen kelerin birbirine ne kadar yakn olabileceini kontrol eder. rnein, 10 piksel olarak ayarlanrsa, en yakn iki ke arasndaki mesafe en az 10 piksel olur.
### corners: kt olarak dndrlen kelerin konumlarn ieren dizi.
### mask: stee bal olarak kullanlan bir maske. Bu maske, ke tespiti yaplacak pikselleri belirlemek iin kullanlr.
### blockSize: Gradyanlarn hesapland piksel blounun boyutu. Genellikle 3x3 boyutunda bir deer kullanlr.
### useHarrisDetector: Eer True olarak ayarlanrsa, Harris ke alglama yntemi kullanlr. False olarak ayarlandnda, Shi-Tomasi ke alglama yntemi kullanlr.
### k: Harris ke alglama ynteminde kullanlan serberst parametre. Eer useHarrisDetector True olarak ayarlanmsa, bu parametre etkinletirilir.

## np.ravel(array, order, Returns)
### Bir numpy dizisini dzletirerek, tek boyutlu bir diziye dntrr. Bu ilem, ok boyutlu dizilerdeki tm eleri tek bir dzletirilmi dizide birletirmek iin kullanlr.
### array: Dzletirilmek istenilen dizi veya matris.
### order: stee bal parametre. Varsaylan olarak "C" deerini alr. Bu deer, dzletirilmi diziye verilerin "satr baznda" kopyalanmas anlamna gelir. "F" sralamas ise, verilerin "stun baznda" kopyalanmas anlamna gelir.
### Returns: Dzletirilmi tek boyutlu bir numpy dizisi dndrr.
