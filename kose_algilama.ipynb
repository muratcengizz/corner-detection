```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(filename="images/sudoku.jpg", flags=0)
img = np.float32(img)
plt.imshow(X=img, cmap="gray");
```

```python
# Harris Corner Detection
harris = cv2.cornerHarris(src=img, blockSize=3, ksize=3, k=0.04)
plt.imshow(X=harris, cmap="gray");
```

```python
# Shi Tomsai Detection
corners = cv2.goodFeaturesToTrack(image=img, maxCorners=100, qualityLevel=0.01, minDistance=10)
corners = np.int64(corners)

for element in corners:
    x, y = element.ravel()
    cv2.circle(img=img,
               center=(x, y),
               radius=3,
               color=(125, 125, 125),
               thickness=cv2.FILLED)
plt.imshow(X=img, cmap="gray");
```

## cv2.cornerHarris(src, blockSize, ksize, k, dst, borderType)
### Harris köşe algılama yöntemini kullanarak bir görüntüdeki köşeleri tespit etmek için kullanılan bir OpenCV fonksiyonudur. Bu yöntem, köşelerin kontrast, yoğunluk ve doğrultu gibi özelliklerine dayanarak köşeleri belirler.
### src: Giriş görüntüsü
### blockSize: Köşe tespiti için kullanılacak piksel bloğunun boyutu. Bu blok içindeki gradyanların hesaplanacağı alanı belirler. Genellikle 2x2 veya 3x3 boyutunda bir değer kullanılır.
### ksize: Sobel gradyan hesaplama için kullanılacak Sobel çekirdeğinin boyutu. Genellikle 3x3 boyutunda bir değer kullanılır.
### k: Harris köşe algılama fonksiyonunda kullanılan serbest parametre. Bu değer, köşe noktalarının belirlenmesinde önemli bir rol oynar. Tipi olarak 0.04 ile 0.06 arasında bir değer kullanılır.

## cv2.goodFeaturesToTrack(image, maxCorners, qualityLevel, minDistance, corners, mask, blocksize, useHarrisDetector, k)
### Bir görüntüdeki iyi takip edilebilir özelliklerin (genellikle köşelerin) tespit edilmesi için kullanılan bir OpenCV fonksiyonudur. Bu fonksiyon, Shi-Tomasi köşe algılama yöntemini kullanır.
### image: Giriş görüntüsü.
### maxCorners: Tespit edilecek en fazla köşe sayısı, Eğer tespit edilen köşe sayısı bu değeri aşarsa, en güçlü köşeler seçilir.
### qualityLevel: Köşe niteliğini belirleyen eşik değeri. Bu değer, köşelerin ne kadar iyi olduğunu kontrol eder. Tipik olarak 0.01 ile 0.1 arasında bir değer kıllanır.
### minDistance: Tespit edilen köşeler arasındaki minimum mesafe. Bu parametre, tespit edilen köşelerin birbirine ne kadar yakın olabileceğini kontrol eder. Örneğin, 10 piksel olarak ayarlanırsa, en yakın iki köşe arasındaki mesafe en az 10 piksel olur.
### corners: Çıktı olarak döndürülen köşelerin konumlarını içeren dizi.
### mask: İsteğe bağlı olarak kullanılan bir maske. Bu maske, köşe tespiti yapılacak pikselleri belirlemek için kullanılır.
### blockSize: Gradyanların hesaplandığı piksel bloğunun boyutu. Genellikle 3x3 boyutunda bir değer kullanılır.
### useHarrisDetector: Eğer True olarak ayarlanırsa, Harris köşe algılama yöntemi kullanılır. False olarak ayarlandığında, Shi-Tomasi köşe algılama yöntemi kullanılır.
### k: Harris köşe algılama yönteminde kullanılan serberst parametre. Eğer useHarrisDetector True olarak ayarlanmışsa, bu parametre etkinleştirilir.

## np.ravel(array, order, Returns)
### Bir numpy dizisini düzleştirerek, tek boyutlu bir diziye dönüştürür. Bu işlem, çok boyutlu dizilerdeki tüm öğeleri tek bir düzleştirilmiş dizide birleştirmek için kullanılır.
### array: Düzleştirilmek istenilen dizi veya matris.
### order: İsteğe bağlı parametre. Varsayılan olarak "C" değerini alır. Bu değer, düzleştirilmiş diziye verilerin "satır bazında" kopyalanması anlamına gelir. "F" sıralaması ise, verilerin "sütun bazında" kopyalanması anlamına gelir.
### Returns: Düzleştirilmiş tek boyutlu bir numpy dizisi döndürür.
